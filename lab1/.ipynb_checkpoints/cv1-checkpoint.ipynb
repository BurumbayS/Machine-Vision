```python
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
im = pltimg.imread('.\\6_26.jpg')
fig = plt.figure(figsize=(15,10))
plt.subplot(221)
plt.imshow(im[:,:,0])
plt.subplot(222)
plt.imshow(im[:,:,1])
plt.subplot(223)
plt.imshow(im[:,:,2])
plt.subplot(224)
plt.imshow(im)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
im = pltimg.imread('6_26.jpg')
temp = im[:,:,0]
im[:,:,0] = im[:,:,2]
im[:,:,2] = temp
plt.imshow(im)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
im = pltimg.imread('6_26.jpg')
temp = im[:,:,1]
im[:,:,1] = im[:,:,2]
im[:,:,2] = temp
plt.imshow(im)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
im = pltimg.imread('6_26.jpg')
temp = im[:,:,0]
im[:,:,0] = im[:,:,2]
im[:,:,2] = im[:,:,1]
im[:,:,1] = temp
plt.imshow(im)
plt.show()
```

```python
im = pltimg.imread('.\\6_26.jpg')
fig = plt.figure(figsize=(15,10))
cat = im[300:600, 900:1200]
plt.subplot(221)
plt.imshow(im)
plt.subplot(222)
plt.imshow(cat)
plt.subplot(223)
plt.imshow(im[0:400, 0:500])
plt.subplot(224)
im[600:900, 900:1200] = cat
plt.imshow(im)
plt.show()
```
